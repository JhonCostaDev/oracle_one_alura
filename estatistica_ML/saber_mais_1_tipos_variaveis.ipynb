# Para saber mais: Tipos de variveis

As variveis so caractersticas ou atributos que coletamos em uma base de dados. Elas podem ser classificadas em dois grandes tipos: variveis qualitativas e variveis quantitativas.




Entender esses tipos e suas subdivises  de extrema importncia para a anlise exploratria e extrao de informaes relevantes com os dados. Vamos explorar cada tipo presente na tabela abaixo que possui os dados sobre a condio e desempenho de jogadores de um time de futebol:

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
jogadores = 'desempenho_jogadores.csv'
import pandas as pd

dados_jogadores = pd.read_csv(jogadores, sep=';')
dados_jogadores.drop(columns='Unnamed: 10', inplace=True)
dados_jogadores
```

Agora que sabemos os dados que possumos, vamos analisar quais deles so qualitativos e quantitativos, e tambm como esto divididos dentro do seu prprio tipo.

### Variveis qualitativas (categricas)
As variveis qualitativas representam caractersticas que no podem ser medidas numericamente, mas podem ser categorizadas, ou seja, representam uma classificao. Elas podem ser subdivididas em:

1. `Variveis qualitativas nominais`: classifica os elementos em classes ou categorias em relao ao tipo do dado representado. Ela no possui uma relao de grandeza ou ordem dos dados. Exemplo:

`Jogador: representando o nome do atleta (ex.: Rodrigo, Pedro)`.
`Posio: a posio em que o jogador ocupa (ex.: GOL, ZAG)`

2. `Variveis qualitativas ordinais`: tambm classifica os elementos em classes ou categorias, porm existe uma ordem natural ou hierrquica entre elas. Exemplo:

`Condio fsica`: a condio fsica atual que o atleta se encontra (ex: Boa, Regular, Ruim)
`Situao`: situao atual do atleta no time (ex: Titular, Reserva, Lesionado).

    Observao: a depender dos tipos dos dados,  possvel que uma varivel qualitativa possa ser representada por nmeros. Por exemplo, uma coluna que representa uma caracterstica em que 0 simboliza No e 1 simboliza Sim. Os nmeros aqui servem para resumir a representao dos dados e no devem ser utilizados para clculo, por exemplo, de mdia.



### Variveis quantitativas (numricas)
As variveis quantitativas representam medidas numricas ou quantidades observveis que podem ser expressas em nmeros. Elas so resultantes de contagens (conjunto finito de valores) ou mensuraes (conjunto infinito de valores). Elas so subdivididas em:

1. `Variveis quantitativas discretas`: quando assumem valores inteiros e contveis. Exemplo:

`Jogos`: nmero de partidas jogadas.
`Gols e Assistncias:` quantidade total de gols e assistncias.
`CA (Cartes Amarelos) e CV (Cartes Vermelhos)`: nmero de cartes recebidos.

2. `Variveis quantitativas contnuas`: podem assumir qualquer valor dentro de um intervalo especfico e geralmente so expressos como nmeros reais. Exemplo:

`Mdia de km`: distncia mdia percorrida em jogos (Ex: 9.5 km, 7.8 km).

Observe como conseguimos descrever como cada varivel se comporta e o que ela apresenta. Com isso fica mais fcil saber definir o que fazer com aquele dado e qual tipo de anlise recorrer. A identificao correta dos tipos de variveis  de suma importncia para:

* `Visualizao de dados`: o tipo de varivel determina o tipo de grfico a ser utilizado. Por exemplo, para variveis qualitativas, podemos utilizar tabelas de frequncia e grficos de barras. E para variveis quantitativas, podemos usar histogramas, boxplots e grficos de disperso.
* `Estatstica descritiva`: definindo o tipo de varivel teremos diferentes ferramentas estatsticas para a explorao dos dados. Seja focando em contagem e anlise de dados frequentes ou calculando as medidas de tendncia e disperso dos dados, por exemplo.

* `Modelagem`: Alguns algoritmos funcionam melhor com variveis numricas, enquanto outros podem incorporar informaes categricas. Saber como tratar cada uma delas aumenta o potencial, por exemplo, de um modelo de aprendizado de mquina.

## identificando dados qualitativos nominais e ordinais em nmeros


Em anlise de dados,  comum encontrar variveis numricas que, na verdade, representam categorias. Por exemplo, cdigos de identificao, classificaes ou respostas de pesquisas codificadas numericamente. Essas variveis so qualitativas e podem ser de dois tipos principais:

* ***Nominais***: Categorias sem uma ordem lgica, como gneros `(1 = Masculino, 2 = Feminino) ou departamentos (1 = TI, 2 = RH, 3 = Marketing)`.

* ***Ordinais***: Categorias com uma ordem natural, como nveis de satisfao `(1 = Insatisfeito, 2 = Neutro, 3 = Satisfeito)`.
Tratar adequadamente essas variveis  essencial para evitar erros, como realizar operaes matemticas em dados que representam categorias. Em Python, podemos usar a biblioteca pandas para tratar essas variveis corretamente, especialmente com o tipo Categorical para variveis ordinais.

### O que  um Categorical?
Em Python, o Categorical do pandas  usado para representar variveis categricas. Ele permite armazenar as categorias de forma eficiente e, no caso de variveis ordinais, definir uma ordem especfica para elas. Isso facilita a manipulao e anlise dos dados.

#### Exemplo:

Uma empresa realiza uma pesquisa de satisfao com clientes e coleta os seguintes dados:

* `cliente_id`: Nmero de identificao do cliente (nominal).
* `nivel_satisfacao`: Respostas codificadas como 1 (Insatisfeito), 2 (Neutro), 3 (Satisfeito) (ordinal).

O objetivo  transformar essas variveis corretamente utilizando Categorical e visualizar a alterao. Podemos fazer isso da seguinte forma:

```python
# DataFrame de exemplo
dados_loja = pd.DataFrame({
    'cliente_id': [101, 102, 103, 104, 105],
    'nivel_satisfacao': [3, 2, 1, 3, 2]
})
dados_loja
```

Convertendo a coluna nvel de satisfao em categorica ordinal

```python
categorias_ordenadas = sorted(dados_loja['nivel_satisfacao'].unique())

dados_loja['nivel_satisfacao'] = pd.Categorical(
    dados_loja['nivel_satisfacao'],
    categories=categorias_ordenadas,
    ordered=True
)
dados_loja['nivel_satisfacao'].dtype
```

```python
dados_loja['nivel_satisfacao'].unique()
```

Nomeando as categorias

```python
# Dicionrio com o valor atual e o novo valor
avaliacao_labels = {
    1: 'Insatisfeito', 
    2: 'Neutro', 
    3: 'Satisfeito'
}

# Aplicando a mudana
dados_loja['categoria_satisfacao'] = dados_loja['nivel_satisfacao'].map(avaliacao_labels)

dados_loja
```

Ordenando os dados pela coluna categoria de satisfao

```python
dados_loja.sort_values('categoria_satisfacao', ascending=False)
```

Note que ordenamos os dados de forma decrescente pela categoria_satisfacao, e o Categorical reconheceu a ordem definida nas categorias (de Satisfeito a Insatisfeito) com base no parmetro `ordered=True`.

O Categorical pode auxiliar em diversas situaes na cincia de dados e anlise com Python, como:

* **Anlise Estatstica**: Para definir variveis categricas e quais estatsticas descritivas aplicar nesses casos.
* **Visualizao de** Dados: Facilita a criao de grficos categricos, como grficos de barras, usando bibliotecas como Matplotlib e Seaborn.
* **Machine Learning**: Prepara variveis categricas para algoritmos que no aceitam dados numricos diretamente.
* **Manipulao de Dados**: Permite ordenaes e agrupamentos em tabelas e relatrios.
Logo, transformar variveis numricas que expressam variveis categricas em Categorical garante maior preciso e facilita a interpretao em anlises qualitativas.
