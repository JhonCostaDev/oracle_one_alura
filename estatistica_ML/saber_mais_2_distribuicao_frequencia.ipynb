# Para saber mais: distribuio de frequncia


Uma distribuio de frequncia  uma organizao de dados em tabelas, que resume as ocorrncias (frequncias) de cada categoria ou classe de dados em um conjunto.  uma tcnica essencial na estatstica descritiva, utilizada tanto para dados qualitativos quanto quantitativos, a fim de visualizar e interpretar padres nos dados.

Cada entrada na tabela contm a contagem de ocorrncias de valores dentro de um grupo ou intervalo especfico, resumindo a distribuio destes valores. Por exemplo, classifica a quantidade de pessoas pelo sexo biolgico, a quantidade de itens de um produto vendido, a faixa de renda de pessoas, etc.

Para construir uma tabela de frequncia, precisamos:

1. Ordenar os valores das variveis em ordem crescente ou decrescente;
2. Determinar a frequncia de cada valor;
3. Agrupar os dados em classes ou intervalos.

**Uma tabela de frequncia pode possuir os seguintes elementos**:

### Frequncia absoluta (f_i)
Representa o nmero de vezes que um determinado valor ou classe aparece em um conjunto de dados.


Onde, *i* representa a classe analisada.

**Exemplo**: Em um conjunto de dados X = {2, 2, 3, 4, 2, 3, 5}:

O valor 2 ocorre 3 vezes, logo `f_2 = 3`.
O valor 3 ocorre 2 vezes, logo `f_3 = 2`.

### Frequncia absoluta acumulada (F_i)
Soma das frequncias absolutas at a categoria/classe n, como na frmula abaixo. Para todas as classes teramos a frequncia total absoluta (F), ou melhor, o nmero total de registros.



Onde, **Fi**  a frequncia acumulada at a i-sima classe.

Exemplo: Usando o mesmo conjunto de dados X = {2, 2, 3, 4, 2, 3, 5}:

`F_2 = f_2 = 3`.
`F_3 = f_2 + f_3 = 3 + 2 = 5`
`F_4 = f_2 + f_3 + f_4 = 3 + 2 + 1 = 6`

### Frequncia relativa (fr_i)
 a proporo ou percentual de ocorrncias de cada categoria/classe em relao ao total de dados.  dada pela razo entre a frequncia absoluta de cada classe e a frequncia total absoluta:



Onde,

`f_i` = frequncia absoluta do valor i.
`F` = frequncia absoluta total ou nmero total de elementos do conjunto de dados.
Para valores percentuais, basta multiplicar cada frequncia relativa por 100.

Exemplo: Com o conjunto de dados X = {2, 2, 3, 4, 2, 3, 5}, temos F = 7:

Para f_2 = 3, a frequncia relativa  de fr_2 = 3/7 = `0.429 ou 42,9%`
Frequncia relativa acumulada (Fr_i)
Soma das frequncias relativas at a categoria/classe n, como na frmula abaixo. Para todas as classes teramos a frequncia total relativa (Fr), ou melhor, 1 ou 100%.



Exemplo: Seguindo o exemplo anterior:
Para o valor 2, Fr_2 = F_2 / n = 3/7 = 0,429
Para o valor 3, Fr_3 = F_3 / n = (f_2 + f_3) / 7 = 5/7 = 0,714

#### Aplicando os conceito 
Uma pesquisa foi realizada para identificar o meio de transporte favorito de trabalhadores em uma grande cidade. As opes so: "Carro", "nibus", "Bicicleta" e "Metr". Vamos calcular a distribuio de frequncia.

```python
import pandas as pd

# Dados das preferncias de transporte
transporte = [
    "Carro", "nibus", "Bicicleta", "Carro", "Metr", "nibus", 
    "Bicicleta", "Carro", "Metr", "Bicicleta", "Carro", "nibus", 
    "Bicicleta", "Carro", "Metr", "Carro", "nibus", "Bicicleta", 
    "Carro", "Metr", "nibus", "Bicicleta", "Metr", "Carro",
    "Bicicleta", "Carro", "Metr", "nibus", "Carro", "Bicicleta",
    "Metr", "nibus", "Carro", "Bicicleta", "nibus", "Metr",
    "Carro", "nibus", "Metr", "Bicicleta", "Carro", "Metr",
    "Bicicleta", "nibus", "Carro", "Metr", "nibus", "Bicicleta"
]
```

Transformando em um DataFrame

```python
dados = pd.DataFrame(transporte, columns=['Meio_Transporte'])
dados.head()
```

Calculando a distribuio de frequncia

```python
# Groupby
tabela_freq_groupBy = dados.groupby('Meio_Transporte').size().reset_index(name='Freq_absoluta')
tabela_freq_groupBy
```

```python
# Value_counts()
tabela_freq_value_counts = dados['Meio_Transporte'].value_counts().reset_index(name='Freq_absoluta')
tabela_freq_value_counts
```

Calculando a frequncia relativa

```python
tabela_freq_groupBy['Freq_relativa'] = round(tabela_freq_value_counts['Freq_absoluta'] / tabela_freq_value_counts['Freq_absoluta'].sum()*100, 2)
tabela_freq_groupBy
```

```python
dados['Meio_Transporte'].value_counts(normalize=True)
tabela_freq_value_counts['Freq_relativa'] = dados['Meio_Transporte'].value_counts(normalize=True).mul(100).values.round(2)
tabela_freq_value_counts
```

Calculando a frequncia Acumulada

```python
tabela_freq_groupBy['Freq_acumulada'] = tabela_freq_groupBy['Freq_absoluta'].cumsum()
tabela_freq_groupBy
```

Calculando a frequncia Relativa Acumulada

```python
tabela_freq_groupBy['Freq_relativa_acumulada'] = tabela_freq_groupBy['Freq_relativa'].cumsum()
tabela_freq_groupBy
```

Ordenando pela frequncia Absoluta

```python
tabela_frequencia = tabela_freq_groupBy.sort_values(by='Freq_absoluta', ascending=False)
tabela_frequencia
```

## Frequncias Relativas em uma tabela Cruzada


Uma tabela cruzada, ou tabela de contingncia,  uma ferramenta utilizada para analisar a relao entre duas variveis qualitativas, exibindo as frequncias absolutas ou relativas de combinaes de categorias, sejam elas dependentes ou independentes.

Ela  amplamente utilizada em anlises exploratrias de dados para identificar padres ou associaes dos quais os grupos so mutuamente exclusivos.

As frequncias relativas de uma tabela cruzada podem ser calculadas de trs formas:

Por coluna: Cada valor na coluna  comparado ao total da respectiva coluna.
Por linha: Cada valor na linha  comparado ao total da respectiva linha.
Por todos os dados: Cada valor  comparado ao total geral.
