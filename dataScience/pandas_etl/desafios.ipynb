### 07 Tratando números em strings
Para fazer um estudo financeiro de distribuição salarial em uma empresa, você precisa transformar os dados referentes aos cargos e deixá-los prontos para um estudo estatístico.

Desse modo, ao ler o arquivo com dados, você encontra o seguinte DataFrame:




Marque a alternativa que torna os dados da coluna quantidade e salario valores numéricos, ou seja, colunas int e float sem a existência do texto que acompanha cada valor.

Para você ter um suporte nos testes, busque utilizar o código abaixo para criar o DataFrame df e encontrar a solução da atividade:



```python
import pandas as pd
import numpy as np

# dados da firma
dados = {
    'cargo': ['Gerencia', 'Coordenação', 'Supervisão', 'Analista', 'Assistente', 'Operação', 'Auxiliar', 'Estágio', 'Assessoria', 'Consultoria'],
    'quantidade': ['2 pessoas', '1 pessoa', '3 pessoas', '4 pessoas', '5 pessoas', '3 pessoas', '2 pessoas', '1 pessoa', '1 pessoa', '1 pessoa'],
    'salario': ['$10.000 reais', '$8.000 reais', '$7.000 reais', '$5.000 reais', '$4.000 reais', '$3.500 reais', '$3.000 reais', '$1.500 reais', '$2.500 reais', '$6.000 reais']
}

# transformando o dicionário em DataFrame
df = pd.DataFrame(dados)

# df é o DataFrame com os dados da firma
df
```

```python
df['quantidade'] = df['quantidade'].apply(lambda x: x.replace('pessoas', '').replace('pessoa', '').strip()).astype(np.int64)
```

```python
df['salario'] = df['salario'].apply(lambda x: x.replace('$', '').replace('reais', '').replace('.', '').strip()).astype(np.float64)
```
